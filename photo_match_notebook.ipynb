```python
excel_file_path = r'C:\Users\User\Documents\match_image_data.xlsx'
df = pd.read_excel(excel_file_path)
print(df.head())
```

```python
IMG_SIZE = (128, 128)

# Assuming `df` is your DataFrame containing paths and labels
original_images = []
opponent_images = []
labels = []

for index, row in df.iterrows():
    try:
        # Load original image
        original_image_path = row['original_image_path']
        original_image = load_img(original_image_path, target_size=IMG_SIZE)
        original_image = img_to_array(original_image) / 255.0  # Normalize
        original_images.append(original_image)

        # Load opponent image
        opponent_image_path = row['opponent_image_path']
        opponent_image = load_img(opponent_image_path, target_size=IMG_SIZE)
        opponent_image = img_to_array(opponent_image) / 255.0  # Normalize
        opponent_images.append(opponent_image)

        # Read label
        label = row['is_infringement']
        labels.append(label)
    except Exception as e:
        print(f"Error loading image for ID {row['ID']}: {e}")

# Convert to numpy arrays
original_images = np.array(original_images)
opponent_images = np.array(opponent_images)
labels = np.array(labels)

print(f"Loaded {len(original_images)} original and {len(opponent_images)} opponent images.")
```

```python

# 
X_train_original, X_test_original, X_train_opponent, X_test_opponent, y_train, y_test = train_test_split(
    original_images, opponent_images, labels, test_size=0.2, random_state=42)

print(f"Training set size: {len(X_train_original)}")
print(f"Test set size: {len(X_test_original)}")
```

```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator

datagen = ImageDataGenerator(
    rotation_range=20,
    width_shift_range=0.2,
    height_shift_range=0.2,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True,
    fill_mode='nearest'
)

# Apply data augmentation to training data
train_datagen = datagen.flow([X_train_original, X_train_opponent], y_train, batch_size=32)
```

```python
def build_siamese_model(input_shape):
    input = Input(shape=input_shape)
    
    # Shared Convolutional Network
    x = Conv2D(32, (3, 3), activation='relu', kernel_regularizer=l2(0.01))(input)
    x = MaxPooling2D((2, 2))(x)
    x = Conv2D(64, (3, 3), activation='relu', kernel_regularizer=l2(0.01))(x)
    x = MaxPooling2D((2, 2))(x)
    x = Conv2D(128, (3, 3), activation='relu', kernel_regularizer=l2(0.01))(x)
    x = MaxPooling2D((2, 2))(x)
    x = Flatten()(x)
    x = Dense(256, activation='relu', kernel_regularizer=l2(0.01))(x)
    x = Dropout(0.5)(x)  # Add dropout layer
    x = Dense(128, activation='relu', kernel_regularizer=l2(0.01))(x)
    
    return Model(input, x)
```

```python
def build_siamese_model(input_shape):
    inputs = Input(input_shape)
    x = Conv2D(32, (3, 3), activation='relu')(inputs)
    x = BatchNormalization()(x)
    x = MaxPooling2D()(x)
    x = Conv2D(64, (3, 3), activation='relu')(x)
    x = BatchNormalization()(x)
    x = MaxPooling2D()(x)
    x = Conv2D(128, (3, 3), activation='relu')(x)
    x = BatchNormalization()(x)
    x = MaxPooling2D()(x)
    x = Flatten()(x)
    x = Dense(128, activation='relu', kernel_regularizer=l2(0.01))(x)
    x = Dropout(0.5)(x)
    x = Dense(64, activation='relu', kernel_regularizer=l2(0.01))(x)
    return Model(inputs, x)
```

```python



def euclidean_distance(vectors):
    (featsA, featsB) = vectors
    sum_squared = K.sum(K.square(featsA - featsB), axis=1, keepdims=True)
    return K.sqrt(K.maximum(sum_squared, K.epsilon()))

input_shape = (128, 128, 3)

# Build the twin networks
base_model = build_siamese_model(input_shape)

# Input layers for the two images
imgA = Input(shape=input_shape)
imgB = Input(shape=input_shape)

# Feature extraction
featsA = base_model(imgA)
featsB = base_model(imgB)

# Calculate the Euclidean distance between the features
distance = Lambda(euclidean_distance)([featsA, featsB])

# The output layer
output = Dense(1, activation="sigmoid")(distance)

# The Siamese Network model
model = Model(inputs=[imgA, imgB], outputs=output)


model.compile(loss="binary_crossentropy", optimizer=RMSprop(learning_rate=0.0001), metrics=["accuracy"])

model.summary()
```

```python
history = model.fit(
    [X_train_original, X_train_opponent], y_train,
    validation_data=([X_test_original, X_test_opponent], y_test),
    batch_size=32,
    epochs=3  # Increase the number of epochs
)
```
